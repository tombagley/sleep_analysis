```python
#import dependencies
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd
import scipy.stats as st
import scipy.stats as stats
```

```python
#Read in dataset
df = pd.read_csv('../datasets/Sleep_health_and_lifestyle_dataset.csv')
df
```

```python


# Create a scatter plot of the data

plt.scatter(df['Age'], df['Sleep Duration'], label='Data Points')

# Calculate the linear regression line using linregress
slope, intercept, r_value, p_value, std_err = linregress(df['Age'], df['Sleep Duration'])

# Plot the linear regression line
plt.plot(df['Age'], slope * df['Age'] + intercept, color='red', label='Linear Regression Line')

# Add the linear regression equation as a text annotation on the plot
equation = f'y = {slope:.2f}x + {intercept:.2f}'
plt.text(0.5, 0.9, equation, ha='center', va='center', transform=plt.gca().transAxes, color='red')

plt.xlabel('Age')
plt.ylabel('Sleep Duration')
plt.title('Linear Regression of Age vs Sleep Duration')
plt.savefig("figures/age_sleep.png")

plt.show()
correlation = st.pearsonr(df['Age'], df['Sleep Duration'])
print(f"The Pearson correlation coefficient between both factors is {round(correlation[0],2)}")
print(f"The r-squared is: {r_value**2}")
print("The p-value is:", p_value)


#Used Xpert Learning Assistance
```
