```python
#import data 
import pandas as pd
df=pd.read_csv("../datasets/Sleep_health_and_lifestyle_dataset.csv")
df.head()
```

```python
#get number of rows
num_rows = len(df)
len(df)
```

```python
# List of occupations
occupations_list = df['Occupation'].unique()

# Create a DataFrame from the list
List_of_occupations_df = pd.DataFrame(occupations_list, columns=['Occupation'])

# Display the DataFrame
print(df)
```

```python
#average age of partcipant 
average_age=df['Age'].mean()
print("Average Age of Participants:", average_age)

#average stress level of partcipant 
average_stress_level=df["Stress Level"]
print("Average Stress Level of Participants:", average_stress_level)


```

```python
#seperate data for only men
male_df = df[df['Gender'] == 'Male']
male_df.head()
```

```python
#seperate data for women only
female_df = df[df['Gender'] == 'Female']
female_df.head()
```

```python
#average hours women slept
female_sleep_time_average = female_df['Sleep Duration'].mean()

#average age of female partcipant 
female_average_age=female_df["Age"].mean()

#average hours men slept 
male_sleep_time_average = male_df['Sleep Duration'].mean()

#average age of male partcipant 
male_average_age=male_df["Age"].mean()

print("Average Hours Slept by Women:", female_sleep_time_average)
print("Average Age of Female Participants:", female_average_age)
print("Average Hours Slept by Men:", male_sleep_time_average)
print("Average Age of Male Participants:", male_average_age)
```

```python
#occupations 

#Group the data by occupation and calculate the average hours slept for men in each occupation
average_hours_slept_by_occupation_men = male_df.groupby('Occupation')['Sleep Duration'].mean() 

#average stress level for each occupation
average_stress_level_by_occupation_men = male_df.groupby('Occupation')['Stress Level'].mean()  

print("Average Hours Slept by Men in Each Occupation:")
print(average_hours_slept_by_occupation_men)
print("\nAverage Stress Level for Men in Each Occupation:")
print(average_stress_level_by_occupation_men)

```

```python
#average quality of sleep for each occupation
average_sleep_quality_by_occupation_men =male_df.groupby('Occupation')['Quality of Sleep'].mean()

print("\nAverage Quality of Sleep for men in Each Occupation:")
print(average_sleep_quality_by_occupation_men)
```

```python
#Group the data by occupation and calculate the average hours slept for men in each occupation
average_hours_slept_by_occupation_women = female_df.groupby('Occupation')['Sleep Duration'].mean() 

#average stress level for each occupation
average_stress_level_by_occupation_women = female_df.groupby('Occupation')['Stress Level'].mean() 

print("Average Hours Slept by Women in Each Occupation:")
print(average_hours_slept_by_occupation_women)
print("\nAverage Stress Level for Women in Each Occupation:")
print(average_stress_level_by_occupation_women)

```

```python
#average quality of sleep for each occupation
average_sleep_quality_by_occupation_women = female_df.groupby('Occupation')['Quality of Sleep'].mean() 

print("\nAverage Quality of Sleep for Women in Each Occupation:")
print(average_sleep_quality_by_occupation_women)
```
